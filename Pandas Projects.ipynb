```python
# 7. Top 10 profitable products
top10_profitable_products = Sa.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(10)

print(top10_profitable_products)

plt.barh(top10_profitable_products.index, top10_profitable_products.values)

plt.xlabel("Profit")
plt.ylabel("Product Name")
plt.title("Top 10 Profitable Products")
plt.show()
```

```python
#Phase 2: Data Cleaning & Preprocessing
#1. Handle missing values.
df.dropna()
```

```python
#2. Convert date column into datetime format.
df["Order Date"] = pd.to_datetime(df["Order Date"], format="mixed")

df["Month"] = df["Order Date"].dt.month

```

```python
df.dtypes
```

```python
#3. Remove duplicate records.
df.duplicated().sum()
```

```python
#4. Create new columns:
# 1. Total Sales
df["Total_Sales"] = df["Quantity"] * df["Sales"]

# 2. Profit Percentage
df["Profit_Percentage"] = (df["Profit"] / df["Sales"]) * 100

# 3. Month from Order Date
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Month"] = df["Order Date"].dt.month
```

```python
df
```

```python
#Phase 3: Numerical Analysis (NumPy + Pandas)
#1. Calculate total revenue.
# Phase 3: Numerical Analysis
total_revenue = Sa["Sales"].sum()
print("Total Revenue:", total_revenue)
```

```python
#2. Calculate average monthly revenue.
Sa["Order Date"] = pd.to_datetime(Sa["Order Date"])
Sa["Month"] = Sa["Order Date"].dt.month
monthly_revenue = Sa.groupby("Month")["Sales"].sum()
avg_monthly_revenue = monthly_revenue.mean()
print("Average Monthly Revenue:", avg_monthly_revenue)
```

```python
#3. Find maximum and minimum sales.
max_sales = df["Sales"].max()

min_sales = df["Sales"].min()

print(max_sales)

print(min_sales)
```

```python
#4. Find top 10 selling products.
top10_selling_product = Sa.groupby("Product Name")["Sales"].sum().sort_values(ascending = False).head(10)

print(top10_selling_product)
```

```python
#5. Find region-wise sales contribution.
region_sales = Sa.groupby("Region")["Sales"].sum()
print(region_sales)
region_contribution = ((region_sales / region_sales.sum()) * 100).sort_values(ascending = False)
print(region_contribution)
```

```python
#Phase 4: Data Visualization (Matplotlib)
#1. Line Chart – Monthly Sales Trend

plt.plot(monthly_revenue.index , monthly_revenue.values , marker = "o" , color= "g")
plt.title(" Monthly Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()

```

```python
#2. Bar Chart – Top 10 Products by Sales


plt.barh(top10_selling_product.index , top10_selling_product.values , label = "Sales")
plt.title("Top10 Selling Product")
plt.xlabel(" Sales")
plt.ylabel("Product Name")
plt.legend()
plt.show()
```

```python
#3. Pie Chart – Region-wise Revenue Contribution

color = ["pink" , "plum" ,"skyblue" ,"gold" ]
explode = [ 0,0,0.1,0]

plt.pie(region_sales.values ,labels = region_sales.index , colors = color, explode = explode,  autopct = "%1.1f" )
plt.show()
```

```python
#4. Histogram – Sales Distribution
plt.hist(Sa["Sales"], bins=20)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()

```

```python
#5. Scatter Plot – Quantity vs Sales
plt.scatter(Sa["Quantity"], Sa["Sales"])
plt.title("Quantity vs Sales")
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.show()
```
